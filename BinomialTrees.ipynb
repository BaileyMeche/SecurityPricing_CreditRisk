# Binomial Trees

```python
import numpy as np
import pandas as pd
import logging
import plotnine as p9

from assetpricing.instructor import binomial_model
```

```python
logging.basicConfig(level=logging.INFO,format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger()
```

```python
binomial_model(100, 95, 0.2, 0.04, 0, 0.5, -1, n=5)
```

```python
accurate = binomial_model(100, 95, 0.2, 0.04, 0, 0.5, -1, n=1000)
```

```python
vals = np.array([binomial_model(100, 95, 0.2, 0.04, 0, 0.5, -1, n=n) for n in np.arange(1,200)])
```

```python
df = pd.DataFrame({'Steps':np.arange(1,200), 'V':vals})
```

```python
(
    p9.ggplot(df, p9.aes(x='Steps', y='V'))
    + p9.geom_point(color='orange')
)
```

```python
jrvals = np.array([binomial_model(100, 95, 0.2, 0.04, 0, 0.5, -1, n=n, tree_type='JR') for n in np.arange(1,200)])
```

```python
df = pd.DataFrame({'Steps':np.arange(1,200), 'CRR':vals, 'JR':jrvals})
```

```python
(
    p9.ggplot(df, p9.aes(x='Steps', y='JR'))
    + p9.geom_point(color='orange')
)
```

```python
df['Parity'] = 'Even'
p = df['Parity'].copy()
p.loc[2*np.arange(100)] = 'Odd'
df['Parity'] = p
```

```python
(
    p9.ggplot(df, p9.aes(x='Steps', y='JR', color='Parity'))
    + p9.geom_point(
    )
)
```

```python
errors = np.abs(df['JR']-accurate)
df['JRERROR'] = errors
```

```python
(
    p9.ggplot(df, p9.aes(x='Steps', y='JRERROR'))
    + p9.geom_point(
    )
)
```

```python
(
    p9.ggplot(df, p9.aes(x='Steps', y='JRERROR'))
    + p9.geom_point()
    + p9.scale_y_log10()+ p9.scale_x_log10()
)
```
